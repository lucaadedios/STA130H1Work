__Summary on Population Parameters and Sample Statistics in Hypothesis Testing__

In hypothesis testing, it's important to understand the difference between sample statistics and population parameters.

Definitions:

__Individual Data Points__ These are the specific values collected from a sample.
__Sample Average:__ This is the average of the sample values, which summarizes the sample.
__Population Mean:__ This is the true average of the entire population that we want to estimate.
__Hypothesized Population Mean:__ This is the average value we assume under the null hypothesis, representing our starting assumption about the population mean.

Key Concept:

The outcomes of hypothesis tests refer to the population mean, not the sample average. Even though we use sample data to perform the test, the conclusion is about the true average of the population, rather than just the average from our sample.

Null Hypothesis:

The null hypothesis typically states that the population mean is equal to the hypothesized value. The goal of the hypothesis test is to determine whether there is enough evidence to reject this assumption.

## Context: Güntürkün (2003) recorded how kissing couples tilt their heads. 80 out of 124 couples, or 64.5% tilted their heads to the right. Simulate a p-value using a "50/50 coin-flipping" model for the assumption of the null hypothesis  $𝐻_{0}$ that the population of humans don't have left or right head tilt tendencies when kissing, and use the table below to determine the level of evidence we have against  $𝐻_{0}$.

- We are calculating the p-value given the null hypothesis that the population has no preference in kissing, which is set at 0.5. We will calculate the p-value based on the sample proportion of 64.5% (or 0.645).

- To do this, we first need to create a sample distribution by simulating coin flips. We will derive a sample distribution of the proportion of times the coin lands on heads (which can be either heads or tails). We have a total of 124 events, with the coin landing heads up 80 times. We will conduct 10,000 simulations of flipping a coin 124 times and count how many times the coin lands on heads 80 times or more. This value will represent the p-value.

```python
import numpy as np

# Parameters
n = 124  # Total couples
observed_heads = 80  # Couples tilting right
n_simulations = 10000  # Number of simulations

# Simulating coin flips (0 for left, 1 for right)
simulated_heads = np.random.binomial(n, 0.5, n_simulations)

# Calculating p-value
p_value = np.mean(simulated_heads >= observed_heads)

print(f'Simulated p-value: {p_value}')
```

As you can see, we used a binomial distribution because we are dealing with a finite number of events, in contrast to a normal distribution, which involves an infinite number of events. The p-value is 0.0007, which, based on the table, provides very strong evidence to reject the null hypothesis. This means that out of those 10,000 simulations, there was a 0.07% chance of landing on heads 80 times or more, which is very rare and calls into question whether the null hypothesis is true.

## Changing Null Hypothesis

To change the hypothesis test from a two-sided to a one-sided approach, we modified the null hypothesis to state that the vaccine does not improve health scores (e.g., "mu is less than or equal to mu0"). The alternative hypothesis was adjusted to indicate that the vaccine improves health scores (e.g., "mu is greater than mu0"). This change eliminates the potential for detecting effects in both directions, resulting in a single critical region located in the upper tail of the distribution, which focuses solely on the possibility of improvement. Consequently, the p-value for the one-sided test is calculated using only this upper tail, generally leading to a smaller p-value compared to a two-sided test, as we are exclusively interested in observing an improvement rather than any change in health scores. This adjustment enhances the specificity of the hypothesis test by concentrating on whether the vaccine has a positive effect rather than merely any effect.

- When simulating, we ensure that the random improvements assume the null hypothesis is true. To achieve this, we randomize the sign of the differences using the code provided below. Unlike bootstrapping, we use simulated data to create a null distribution, which represents what the values and probabilities would look like if the null hypothesis were true.

Here’s the adjusted code from "Demo II of the Week 5 TUT," with in-depth multiline comments explaining each part:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample patient data (for illustration purposes)
# You can replace this sample data with your actual dataset.
patient_data = pd.DataFrame({
    'HealthScoreChange': [0.5, -0.2, 0.3, 0.1, -0.4, 0.6, -0.1, 0.4, -0.3, 0.2]  # Example values
})

number_of_simulations = 10000  # Number of simulations to estimate the p-value.

"""
1. Define the observed statistic.
   The observed statistic is calculated as the proportion of patients whose health scores improved.
"""
observed_statistic = (patient_data['HealthScoreChange'] > 0).mean()

"""
2. Simulate under the null hypothesis.
   Initialize an array to store the results of each simulation, representing the proportion of improvements.
"""
IncreaseProportionSimulations_underH0random = np.zeros(number_of_simulations)

for i in range(number_of_simulations):
    """
    Randomly assign improvements under the null hypothesis.
    Each patient's outcome is simulated as either improved (1) or not improved (0).
    """
    random_improvement = np.random.choice([0, 1], size=len(patient_data), replace=True)
    
    """
    Calculate the proportion of patients who improved in this simulation and store it.
    """
    IncreaseProportionSimulations_underH0random[i] = random_improvement.mean()

"""
3. Calculate the p-value for a one-sided test.
   The p-value is determined by counting how many simulated proportions are greater than or equal to the observed statistic.
"""
p_value = (IncreaseProportionSimulations_underH0random >= observed_statistic).sum() / number_of_simulations

print("p-value for one-sided test:", p_value)  # Display the calculated p-value.

"""
4. Visualization of results.
   Plotting the distribution of simulated proportions helps in interpreting the results visually.
"""
plt.figure(figsize=(10, 6))
sns.histplot(IncreaseProportionSimulations_underH0random, bins=30, color='skyblue', kde=True)
plt.axvline(observed_statistic, color='red', linestyle='--', label='Observed Statistic')
plt.title('Distribution of Simulated Proportions Under Null Hypothesis')
plt.xlabel('Proportion of Patients Improved')
plt.ylabel('Frequency')
plt.legend()
plt.show()  # Display the plot.
```

## Explanation of Parts of Graph

- The blue bars represent the histogram of the simulated proportions under the null hypothesis. Each bar corresponds to a range of proportion values, and its height indicates how many simulations resulted in proportions falling within that range. This gives us an idea of the distribution of simulated data points and how frequently certain proportions of improvement occurred.

- The blue squiggly line overlaid on the histogram represents the kernel density estimate (KDE). This smooth curve provides a continuous estimate of the probability density function of the simulated proportions, allowing us to see the overall trend in the data. It helps identify the areas where the proportions are more densely populated and provides a clearer view of the distribution.

- The red dashed line indicates the observed statistic, which is the proportion of patients whose health scores improved based on your actual data. This line helps you visually compare the observed data to the simulated distribution under the null hypothesis. If the red dashed line falls to the right of the bulk of the blue bars, it suggests that the observed proportion of improvements is higher than what would typically be expected under the null hypothesis.

## Interpretation of the Hypothesis Test

By transitioning to a one-tailed hypothesis test, we focus solely on the possibility of improvement due to the 
vaccine, rather than any change in health scores. This change in hypothesis affects the interpretation of our results; a smaller p-value indicates stronger evidence against the null hypothesis, emphasizing that the vaccine has a positive effect on health scores.

## Expectation of P-value

In general, we expect the p-value from the one-tailed test to be smaller than that from a two-tailed test. This is because the one-tailed test only considers one direction (improvement) in the distribution, leading to a more concentrated area of rejection compared to the two-tailed test, which divides the significance level between two tails.

## Introduction

- We are applying the story of Fisher's Tea Experiment to the STA130H1 students. In this context, Dr. Fisher and Dr. Bristol were testing whether Dr. Bristol could actually taste the difference between tea poured first before the milk or vice versa. Dr. Fisher conducted a mini-experiment with 8 cups arranged in different orders, and Dr. Bristol was able to correctly identify all 8 drinks based on taste. Based on this story, we will construct an experiment involving 80 students in the class instead of just one participant like Dr. Bristol. Each student will be assigned one cup—either with the milk poured first or the tea poured first—and we will record how many out of the 80 students correctly guessed the pouring order. The 2 key differences is that the sample size is much larger testing out 80 students rather than just 1 which changes the statistical power and reliability of the conclusion. Finally the parameter being tested was tailored towards Dr. Bristol depending on individual sensory, in this case we use STA130H1 as a collective ability to guess correctly.

- Regarding the population and sample, the population of interest consists of all students in the STA130H1 course, including a diverse range of individuals with varying taste preferences.
- The sample is the 80 students who were randomly selected from this larger population, used to make inferences about the population and its parameters. In this case, the parameter analyzed is the proportion of students who correctly guess the difference between the cups.

## Null and Alternative Hypothesis
- The formal version of the null hypothesis tested would be $H_{0}$: P = 0.5. In simpler terms, the null hypothesis assumes that there is no real difference in the students' ability to differentiate based on the order of pouring milk or tea. Any successes would be attributed purely to chance, akin to flipping a coin. Theoretically, this means that 40 out of the 80 students would have guessed correctly if they were merely guessing.

- The alternative Hypothesis is $H_{0}$: P != 0.5, which suggests that the students' ability to distinguish between the two pouring orders is better than what would be expected by random guessing.

## Method of Testing
To do this, we will calculate the p-value, which is the likelihood of observing data as extreme as (or more extreme than) what we observed, assuming that the null hypothesis is true. In this context, we are assessing how likely it is to see the observed statistic—49 out of 80 students guessing correctly (which is 0.6125)—given that there is no difference in the ability to distinguish between the two pouring orders.

I will now explain the code below to generate the p-value accompanied by a visual.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(42)

"""
In this code, we will generate 10,000 simulations of 80 students, assuming a 50% success rate for correctly 
identifying the order of pouring (either milk first or tea first). The values used in the code are simply for 
readability and to clarify the meaning of the numbers. The choice of 10,000 is arbitrary; it represents a 
sufficiently large number of synthetic simulations that allow us to artificially generate data points based on 
the assumption of the null hypothesis, which suggests a success probability of 0.5 (or 50%).
"""
n_trials = 10000 
n_students = 80   
success_prob = 0.5  


simulated_successes = np.random.binomial(n_students, success_prob, n_trials)
"""
We use a binomial distribution to model the number of successful outcomes in a fixed number of trials; in this 
case, there are 80 trials representing the students guessing. For each trial, there are two possible outcomes: the 
student either guesses correctly or incorrectly. This encapsulates the essence of a binomial distribution, which 
deals with scenarios involving two outcomes. For the variable simulated_successes, we create an array containing 
10,000 items, where each item represents the number of correct guesses out of the 80 trials in a simulated 
experiment. Each of these items corresponds to a different outcome based on the specified probability of 
success (in this case, 50%).
"""


observed_successes = 49 
observed_proportion = observed_successes / n_students
"""
Here above we simply just create the observe statistic which was 49/80 as said which is 0.6125.
"""

# Calculate the p-value
p_value = np.mean(simulated_successes / n_students >= observed_proportion)
"""
So what this does is the n_students divides each element in the array of simulated_successes to get the proportion
of students that got it correct out of 80. Then after that for each element it checks if it's greater than or 
equal to the observered proportion value which we discussed is 0.6125. Every time it is because it is a boolean
expression it will return 1, and then 0 if the element proportion is less than the observed proportion. After
the 10,000 trials, the .mean() will accumulated how many times the array return true(1) and sum all that up, then
divide it by the total amount of trials which was 10,000 essentially mimicking how likely we would see the observed
statistic or more extreme values to that.
"""

# Visualization
plt.figure(figsize=(10, 6))
sns.histplot(simulated_successes / n_students, bins=30, kde=True, color='skyblue', stat='density')
plt.axvline(observed_proportion, color='red', linestyle='--', label='Observed Proportion (0.6125)')
plt.title('Simulated Proportion of Correct Identifications under Null Hypothesis')
plt.xlabel('Proportion of Correct Identifications')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Print the p-value
print(f'P-value: {p_value:.4f}')
```

## Explanation of Graph
As you can see, we have both a histogram with various bins that contain the values synthetically distributed onto the distribution, as well as a KDE represented by the smooth blue line, which helps visualize the data in a more continuous manner. The red dashed line indicates the observed statistic and shows its position on the distribution.

## Interpretation of P-Value
Based on the p-value being 0.0294, we would say that, given the null hypothesis is true, we would see this result 2.94% of the time. Additionally, it is compared to alpha, which is typically set at 0.05. Since the p-value (0.0294) is less than alpha (0.05) which is the significance level, we reject the null hypothesis. This indicates that there is statistically significant evidence to suggest that the order of pouring the milk or tea does affect students' ability to correctly guess which was poured first. In other words, the results are unlikely to have occurred by random chance alone, supporting the idea that there is a genuine difference in how the two pouring methods are perceived. Again stated from above, this does not definitively say that the null hypothesis is not a plausible idea only strong evidence as we can only find out if we know the true population parameter. Meaning even after rejecting the null hypothesis, it doesn't definitively prove that the null hypothesis is false; it only suggests that the evidence against it is strong based on the sample data.
