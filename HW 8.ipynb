__Classification Decision Trees:__
- A classification decision tree works by taking in data and separating it into distinct classes by identifying unique features that divide the data into subsets. This process involves using binary questions at decision nodes, where data that meets the criteria moves to the next decision node, while data that doesn't meet the criteria proceeds along a different path, eventually reaching a leaf node that classifies the subset. The purpose of a classification decision tree is to identify patterns or structures within categorized data. It typically uses multiple decision nodes (binary questions or if-else statements) to categorize data into different classes.
- The key difference between a classification decision tree and a regression tree is that in classification, the goal is to categorize data into discrete classes, while in regression, the aim is to predict continuous numerical outcomes, usually by modeling the relationship between the predictor variables and the outcome variable.
- Some real-world problems where they can use this could be classifying if a person is eligible for a scholarship so they have a decision node see if they are under 18 years old(x <= 18) and then if yes it will continue and if no the decision node will go to a leaf node classifying that individual of not being eligible for a scholarship.

__Decision Trees vs Multiple Linear Regression:__
- The process begins at the root node, where the first decision is made based on a feature (e.g., "Is feature \(x_1 > 5\)?").
- Internal nodes (decision nodes) represent binary decisions or questions that split the data further.
- The process continues down the tree with more decisions until the data reaches a leaf node.
- Leaf nodes contain the final class label, which represents the majority class of the data points in that leaf.
- Each leaf node classifies all data within it into a specific category.
- The process is sequential: each decision node narrows down the data by making decisions based on features, eventually reaching a class label.
<br>

- Multiple Linear Regression (MLR) uses a linear combination of all features (variables) to make a single continuous prediction.
- Instead of sequentially splitting data, MLR captures the overall relationship between the features and the target variable in one mathematical formula.
- The model makes a prediction by combining features into a weighted sum, where each feature is multiplied by its corresponding coefficient and then added with an intercept term.
- The result is a continuous value (e.g., predicting house prices, temperatures, etc.).

__Key Comparison:__ 
- Decision Trees predict discrete class labels by sequentially making binary decisions at internal nodes based on feature values, with the final classification made at the leaf node. In contrast, Multiple Linear Regression predicts a continuous value by combining all features into a weighted sum with coefficients, resulting in a single output.


# Chat Summary

### **Classification Decision Trees**:
1. **Root Node**: The process starts with the root node, where the first decision is made based on a feature (e.g., "Is feature \( x_1 > 5 \)?").
2. **Internal/Decision Nodes**: Each internal node (also known as decision nodes) represents a **binary decision** or **question** based on a feature. This splits the data further.
3. **Leaf Nodes**: After a series of binary decisions, the data reaches **leaf nodes**. Each leaf node contains a final **class label**, which classifies all data points within that node.
4. **Sequential Process**: The decision-making process is sequential, where predictions are made based on features at each internal node. The data is split down the tree until it reaches a leaf, where a class label is assigned.

### **Multiple Linear Regression (MLR)**:
1. **Linear Combination of Features**: MLR predicts a **continuous value** by creating a **linear combination** of the features. This means each feature is multiplied by its corresponding coefficient (weight) and summed, with an intercept term added.
   \[
   \hat{y} = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
   \]
2. **Capturing Relationships**: Instead of sequentially splitting data, MLR captures the **overall relationship** between all features and the target variable in one formula.
3. **Continuous Prediction**: The result is a **continuous output** (e.g., predicting house prices, temperatures), rather than a discrete class label.

### **Key Comparison**:
- **Classification Decision Trees** predict **discrete class labels** through a **sequential decision-making process** at each node, ending in a classification at the leaf.
- **Multiple Linear Regression** predicts a **continuous value** by **combining all features** with weighted coefficients in one linear formula.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Step 2: Load and Preprocess the Dataset
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Drop unnecessary columns and clean data
columns_to_drop = ['Weight_oz', 'Width', 'Height']
ab = ab.drop(columns=columns_to_drop).dropna()
ab['Pub year'] = ab['Pub year'].astype(int)
ab['NumPages'] = ab['NumPages'].astype(int)
ab['Hard_or_Paper'] = ab['Hard_or_Paper'].astype('category')

# Step 3: Split Data into Training and Testing Sets
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab, test_size=0.2, random_state=42
)

# Report dataset sizes
print(f"Training data size: {len(ab_reduced_noNaN_train)} observations")
print(f"Testing data size: {len(ab_reduced_noNaN_test)} observations")

# Step 4: Train Decision Tree Classifier
# Prepare features and target
X_train = ab_reduced_noNaN_train[['List Price']]
y_train = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']

# Initialize and train the classifier
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train, y_train)

# Step 5: Visualize the Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True)
plt.title("Decision Tree for Hardcover vs Paperback")
plt.show()
```

- In linear regression, the coefficients represent the direct relationship between each feature and the predicted outcome. Specifically, each coefficient quantifies how much the predicted value will change for a one-unit increase in the corresponding feature, while holding all other variables constant. This linear relationship assumes that the impact of each feature is additive and independent of others. On the other hand, decision trees use feature importances to indicate the relative significance of each feature in splitting the data to reduce impurity (such as Gini impurity or entropy). Unlike linear regression, feature importances in decision trees do not represent a direct, linear relationship with the predicted outcome. Instead, they reflect how much each feature contributes to improving the model's overall predictive accuracy by determining the most informative splits at each node in the tree. This means that decision trees capture more complex, non-linear interactions between features and their contributions to the final prediction. It's like comparing oranges to apples where these are almost completely unrelated things, one is measuring the direct linear relationship, while the other is looking at the most important features for splitting purifying for data splitting in a decision tree.
